```python
from pathlib import Path
```

```python
raw_input_directory = Path("raws/input")
```

```python
for label in raw_input_directory.glob("*.txt"):
    with open(label, 'r') as f:
        lines = [line for line in f if not line.startswith('0 ')]
    with open(label, 'w') as f:
        f.writelines(lines)
```
